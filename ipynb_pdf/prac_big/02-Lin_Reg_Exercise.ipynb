```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
pd.set_option('display.max_columns', None)
```

<p class="question">
    Read the file <code>house_prices.csv</code>
</p>

Original data source: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

```python
##df =  TODO: Complete Code ##
df = pd.read_csv('house_prices.csv', encoding='utf-8-sig')
```

```python
df.head(10)
```

Fill null values with zero:

```python
df = df.fillna(0)
```

<p class='question'>
    <code>X</code> should contain all columns except the last one <code>SalePrice</code>
    and <code>y</code> should contain the values of the column <code>SalePrice</code>
    </p>

```python
##X =  TODO: Complete Code ##
##y =  TODO: Complete Code ##
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

<p class='question'>
    Build a Linear Regression model using <code>sklearn</code> and fit it
    </p>

```python
## TODO: Complete Code ##
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
```

<p class='question'>
    Calculate the mean square error
    </p>

```python
## TODO: Complete Code ##
from sklearn.metrics import mean_squared_error

y_pred = reg.predict(X)
mse = mean_squared_error(y, y_pred)
mse
```
