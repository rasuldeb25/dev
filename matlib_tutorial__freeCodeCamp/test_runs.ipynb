```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels,values)
bars[0].set_hatch("/")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels,values)
bars[0].set_hatch("/")
bars[1].set_hatch("O")
bars[2].set_hatch("*")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1,4,2]

patterns = ["/", "O", "*"]
bars = plt.bar(labels,values)
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

gas = pd.read_csv("gas_prices.csv")
plt.figure(figsize=(8,5))

plt.title("Gas prices over time in USD")

plt.plot(gas.Year, gas.USA, label="The US", color="blue", linestyle="-", marker=".") #-> instead of linestyle/marker/color we can use 'b.-'
plt.plot(gas.Year, gas.Canada, label="Canada",color="red", linestyle="-", marker=".")
plt.plot(gas.Year, gas["South Korea"], label="North Korea", color="green", linestyle="-", marker=".")
plt.plot(gas.Year, gas.Australia, label="Australia", color="orange", linestyle="-", marker="." )

plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

gas = pd.read_csv("gas_prices.csv")
plt.figure(figsize=(12,8))

plt.title("Gas prices over time in USD")

for country in gas:
    if country != "Year":
        plt.plot(gas.Year, gas[country], marker=".", label=country)

plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4, .2, 0, 0, .4)

plt.title('Weight Distribution of FIFA Players', weight="bold")
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use = "default"

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]

labels = ["FC Barcelone", "Real Madrid", "New England Revolution"]

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={"linewidth": 2})


for box in boxes["boxes"]:
    box.set(color="#4286f4",
            linewidth = 2)
    box.set(facecolor="#e0e0e0")
plt.title("Professional Soccer Team Comparasion")
plt.ylabel("Fifa overall rating")

plt.show()
```
