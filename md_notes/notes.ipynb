**Gromacs**

#removing HOH<br>
    grep -v HOH .pdb > .pdb //doing so is not always right way<br>
#box defining<br>
    gmx editconf -f .gro -o .gro -c -d 1.0 -bt cubic<br>
#solvating<br>
    gmx solvate -cp .gro -cs spc216.gro -o .gro -p topol.top<br>
#running<br>
    gmx grompp -f .mdp -c .gro -p .top -r .gro -o name.tpr<br>
        gmx mdrun -deffnm name<br>
#analysis<br>
    gmx trjconv -s .tpr -f .xtc -o .xtc -pbc mol/cluster/nojump -center<br>
