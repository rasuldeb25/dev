Lesson 2: 1D Kinematics and Loops

1st prblem:
*A cart is on a track at x=0 and at t=0s
*The cart is moving with an x-velocity of 0.45m/s
*Where is the cart at t=1.5s? 

```python
#solution of 1st problem
x = 0
v = 0.45
t = 0
dt = 0.25

while t < 1.5:
    print("t = ", t)
    x = x + v*dt
    t = t + dt
print("x= ", x, " m")
```

2nd problem:
*A cart starts at x = 0m, and t=0s, with the velocity of 0.45m/s
*The cart has a acceleration of 0.2m/s^2
*Where is the cart after 1.5s?

```python
#solution of 2nd problem
x = 0
v = 0.45
t = 0
dt = .25
a = 0.2

while t < 1.5:
    v = v + a*dt
    x = x + v*dt
    t = t + dt

print("x= ", x, " m")
```
