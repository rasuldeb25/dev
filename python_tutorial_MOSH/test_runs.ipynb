```python
#Parameters/positional arguments
def greet_user(first_name, last_name): #we can add parameters inside of the functions 
    #difference betwee parametres and arguments:
    #paramenteres are the palceholders that we define in our function to recieve information
    #arguments are the actual pieces of information passed to the functions
    print(f"Hi {first_name} {last_name}!")
    print("Welcome abroad")


print("Start")
greet_user("John", "Smith")
greet_user("Mary", "Tedd") #these are positional arguments
print("End")
```

```python
#Keyword arguments
def greet_user(first_name, last_name):
    print(f"Hi {first_name} {last_name}!")
    print("Welcome abroad")


print("Start")
greet_user(last_name="Smith", first_name="John")
print("End")
```

```python
#Inheritance
class Mammal:
    def walk(self):
        print("walk")

class Dog(Mammal):
   def bark(self):
       print("Dog is barking")

class Cat(Mammal):
    def meow(self):
        print("Cat is meowing")

dog1 = Dog()
dog1.walk()
dog1.bark()

cat1 = Cat()
cat1.walk()
cat1.meow()
```
