data->model->predic

We split the dataset up to X and Y: X is everything we are using to make a predictions, Y is the prediction that we are interested in making

In the case of house price prediction, Y dateset contains `house prices`, while X contains general information about the house (sq²,how big is the garden etc.)

data (X,y)->scale->model(Y is used to predict)->predict

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
mod = LinearRegression().fit(X,y)
pipe = Pipeline([
    ("scale", StandardScaler()),
    ("model", KNeighborsRegressor(n_neighbors=1))
])
```

```python
pipe.fit(X,y)
```

```python
pred = pipe.predict(X)
```

```python
plt.scatter(pred,y)
```
