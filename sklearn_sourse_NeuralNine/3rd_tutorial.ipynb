```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X,y = data.data, data.target
```

if you overfit you training set, you basically memorize it which is not good, so what we do in Ml is we train the model on one portion of the data and we evaluate on the other/unsees portion of the data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2) #test_size is 0.2 -> means 20% of data being used for testing, rest will be used for traning
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for train_ndx, test_idx in split.split(X,y):
    X_train, X_test = X[train_ndx], X[test_idx]
    y_train, y_test= y[train_ndx], y[test_idx]
```
